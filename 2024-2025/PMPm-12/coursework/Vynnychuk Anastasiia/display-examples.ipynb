```python
import cv2 as cv
from ultralytics import YOLO
import matplotlib.pyplot as plt
import os
from os import path
```

```python
model = YOLO('./runs/detect/train40epochs/weights/best.pt')
```

```python
examples_dir = './data/examples'
inputs = [f'{examples_dir}/{file_name}' for file_name in os.listdir(examples_dir) if file_name.endswith('.jpg')]

results = model(inputs)
```

```python
for (file_path, result) in zip(inputs, results):
    img_with_bb = result.orig_img.copy()
    for box in result.boxes:
        img_with_bb = cv.rectangle(img_with_bb, box.xyxy[0,:2].to(int).tolist(), box.xyxy[0,-2:].to(int).tolist(), (0, 0, 255), thickness=2)

    file_name = file_path.split('/')[-1]
    cv.imwrite(f'./data/examples/results/{file_name}', img_with_bb)
```
